```python
import numpy as np
from kithcmb import kitaevhoneycomb
from matplotlib import pyplot as plt

J = [1,1,1]
K = 0.25
```

# Site labelling

It's useful to have two different labellings of the lattice.

```python
Lrows,Lcols = 4,4

ux = np.ones((Lrows,Lcols))
uy = np.ones((Lrows,Lcols))
uz = np.ones((Lrows,Lcols))

kh_sys = kitaevhoneycomb.kitaevhoneycomb(J=J,K=K,ux=ux,uy=uy,uz=uz)
```

## One-dimensional site labelling

This is the indexing of the lattice sites used in the the coupling matrix (A_ij) of the Hamiltonian.  
The indexes i and j run from 0 to (N-1) and the image below shows how these indexes relate to the lattice geometry.

```python
kh_sys.draw_one_d_labelling(figsizeboost=2)
```

## Resolved lattice labelling (row, column, sublattice)
This labelling makes it easy to write down where the couplings between sites should be.

```python
kh_sys.draw_resolved_labelling(figsizeboost=2.)
```

The `kitaevhoneycomb` objects have two internal arrays that convert between these two different labellings:  
`resolved_index_to_one_d[row,col,bw] = i` and `one_d_index_to_resolved[i] = [row,col,bw]` 

This is used in the `make_A` function to vectorise the construction of the Hamiltonian matrix A_ij.

There is another array in the `kitaevhoneycomb` objects called `real_space_positions[i] = [x,y]` that  
gives the positions in real space of the lattice sites and is used in the drawing functions:

```python
fig,ax = plt.subplots(figsize=(8,6))
ax.set_aspect('equal')
plt.axis('off')

# identify 'blue' sublattice sites
sites = (kh_sys.one_d_index_to_resolved[:,2]==0)
ax.scatter(kh_sys.real_space_positions[sites,0],kh_sys.real_space_positions[sites,1],c='C0',s=100)

# identify 'orange' sublattice sites
sites = (kh_sys.one_d_index_to_resolved[:,2]==1)
ax.scatter(kh_sys.real_space_positions[sites,0],kh_sys.real_space_positions[sites,1],c='C1',s=100)

plt.show()
```

# Encoding the gauge fields

```python
Lrows,Lcols = 4,4
```

By zeroing all of the link variables, apart from the ones associated with a single unit cell you can see how the `u` arrays relate to the lattice links

```python
ux = np.zeros((Lrows,Lcols))
uy = np.zeros((Lrows,Lcols))
uz = np.zeros((Lrows,Lcols))

ux[0,0] = 1
uy[0,0] = 1
uz[0,0] = 1

kh_sys = kitaevhoneycomb.kitaevhoneycomb(J=J,K=K,ux=ux,uy=uy,uz=uz)

kh_sys.draw_system(figsizeboost=2)
```

The colouring of the links is by comparison with the canonical version of the no-vortex sector that we choose, where all the `ux`, `uy`, `uz` are 1.  
If `u` values are flipped relative to this they are drawn in red.

```python
ux = np.ones((Lrows,Lcols))
uy = np.ones((Lrows,Lcols))
uz = np.ones((Lrows,Lcols))

uz[uz.shape[0]//2,:] = -1

kh_sys = kitaevhoneycomb.kitaevhoneycomb(J=J,K=K,ux=ux,uy=uy,uz=uz)

kh_sys.draw_system(figsizeboost=2)
```

Vortices are drawn with black dots
